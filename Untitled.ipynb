```python
import wikipediaapi

def get_book_page_names(q):
    headers = {
        'User-Agent': 'my-python-script/1.0 (your-email@example.com)' 
    }
    
    search_url = f"https://en.wikipedia.org/w/api.php?action=query&list=search&srsearch={q}&utf8=&format=json"
    search_response = requests.get(search_url, headers=headers).json()
    try:
     
        page_titles = [search_response['query']['search'][i]['title'] for i in range(len(search_response['query']['search']))][:5]
        
    except IndexError:
        print(q)
        page_titles = [np.NaN]
    
    return page_titles

def get_plot(q):
    wiki_page_titles = get_book_page_names(q)
    wiki_wiki = wikipediaapi.Wikipedia(
        language='en', 
        user_agent="YourAppName/1.0 (your-email@example.com)" 
    )

    for wiki_page_title in wiki_page_titles:
        page = wiki_wiki.page(wiki_page_title)

        def get_section_text(section):
            text = section.text
            for subsection in section.sections:
                text += " " + get_section_text(subsection)
            return text
    
        for plot_indicator in ['Plot', 'Plot summary']:
            plot_section = page.section_by_title(plot_indicator)
            if plot_section:
                return get_section_text(plot_section).strip()
    
    return np.NaN

author = "George Saunders"
title = "Lincoln in the Bardo"
get_plot(f"{title} (novel) by {author}")
```

```python
from ollama import chat
from pydantic import BaseModel, Field
from typing import Literal

class Character(BaseModel):
    name: str
    role: str | None
    age: int | None
    backstory: str | None
    beliefs: str | None
    weaknesses: str | None
    strengths: str | None
    motivations: str | None
    #actantial_role: Literal['Subject', 'Object', 'Helper', 'Opponent', 'Sender', 'Receiver'] = None

class Setting(BaseModel):
    time_periods: list[str] | list
    locations: list[str] | list
    cultural_context: list[str] | list

class Plot(BaseModel):

    #timeline_of_events: list[str] = Field(description = (
    #    "a sequenced list of very brief event descriptions"
    #))
    
    conflict: Literal['person vs. person', 'person vs. self', 'person vs. nature', 'person vs.society'] | list

    story_exposition: str | None
    story_rising_action: str | None
    story_climax: str | None
    story_falling_actions: str | None
    story_denouement: str | None
    
class Theme(BaseModel):
    main_themes: list[str]
    secondary_themes: list[str]
    
class Other(BaseModel):
    main_genres: list[str]
    sub_genres: list[str]
    content_warnings: list[str]
    writing_style: str = None
    point_of_view: Literal['First-person narrator', 'Third-person limited narrator', 'Third-person omniscient narrator'] = None
    
class Elements(BaseModel):
    characters: list[Character]
    setting: Setting
    plot: Plot
    theme: Theme
    other: Other
    
```

```python
%%time
response = chat(
    messages=[
    {"role": "system", "content": f"Return as a JSON object"},
    {"role": "user", "content": text}
    ],
    model='llama3.1:8b',
    format=Elements.model_json_schema(),
    options = {'temperature':0}
)

elements = Elements.model_validate_json(response.message.content)
print(elements)
```
